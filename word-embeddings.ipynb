```python
nltk_stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Helper Function: Normalize Tweet
def normalize_tweet(tweet):
    return tweet.lower()
```

```python
import contractions
def fix_contractions(tweet):
    return contractions.fix(tweet)
```

```python
def remove_noisy_tokens(tweet):
    return re.sub(r'@[a-zA-Z0-9]+|#[a-zA-Z0-9]+|\w+://[^\s]+|\W+|\d+|_+|[^\u0000-\u007f]+', ' ', tweet)
```

```python
def remove_remaining_noisy_tokens(tweet):
    return re.sub(r'\b\w\b|[^\u0000-\u007f]+|_+|\W+', ' ', tweet)
```

```python
def remove_stopwords(tokens):
    return [token for token in tokens if token not in nltk_stopwords]
```

```python
def penn_to_wordnet(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def lemmatize(tokens):
    pos_tags = pos_tag(tokens)
    return [lemmatizer.lemmatize(token, penn_to_wordnet(tag)) for token, tag in pos_tags]
```

```python
def preprocess_tweet(tweet):
    tweet = normalize_tweet(tweet)
    tweet = fix_contractions(tweet) 
    tweet = remove_noisy_tokens(tweet)
    tweet = remove_remaining_noisy_tokens(tweet)
    tokens = word_tokenize(tweet)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    return ' '.join(tokens)
```

```python
def preprocess_data(data):
    with ThreadPoolExecutor(max_workers=os.cpu_count()) as executor:
        data['text'] = list(executor.map(preprocess_tweet, data['text']))
    return data
```

```python
data = preprocess_data(data)
```
